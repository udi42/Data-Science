```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('C:/Users/udith_/Downloads/basket.csv')
```

```python
# Identify the structure of dataset
data.info()
```

```python
# Identify the summary of dataset
data.describe()
```

```python
# To find the null values
data.isna().sum()
```

```python
# Fill null values
data=data.fillna('')
```

```python
data.isna().sum()
```

```python
df=pd.DataFrame(data)
df
```

```python
oht = pd.get_dummies(df)
```

```python
# Apriori
frequent_item_sets = apriori(oht, min_support=0.4, use_colnames=True)
```

```python
# Association rules
association_rules_df = association_rules(frequent_item_sets, metric='confidence', min_threshold=0.7)
association_rules_df
```

```python
# Frequent_item_set
association_rules_df
```

```python
rules = association_rules(frequent_item_sets, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
print(rules.head(5))
```

```python
association_rules_df["antecedents_len"]=association_rules_df['antecedents'].apply(lambda x:len(x))
association_rules_df=association_rules_df[association_rules_df['antecedents_len']>1]
association_rules_df = association_rules_df.sort_values(['confidence','support'], ascending =[False,False])
print(association_rules_df.head(5))
```
