```python
import pandas as pd
import numpy as np
import seaborn as sns
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import nltk
from nltk import tokenize
from nltk.probability import FreqDist
import urllib.request
nltk.download('punkt')
%matplotlib inline
```

```python
textfile = open('C:/Users/udith_/Downloads/Nile.txt', 'r').read()
print(textfile[1000:1500])
```

```python
# Tokenize the text in a file and count the total number of words
from nltk import word_tokenize
words = word_tokenize(textfile)
words
print(f"Total number of words in the file is: {len(words)}")
```

```python
# Calculate and display the ten most common words in the tokenized text
fdist = FreqDist(words)
fdist.most_common(10)
```

```python
# Filter and count alphabetic words in lowercase, excluding punctuation
words_no_punc = []
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())
print(f"Words without punciation is: {len(words_no_punc)}")
```

```python
# Download and print the list of English stopwords using NLTK
nltk.download("stopwords")
from nltk.corpus import stopwords
stopwords_list = stopwords.words("english")
print(stopwords_list)
```

```python
# Remove stopwords and non-alphabetic words, then count the cleaned words
clean_words = []
for word in words_no_punc:
    if word not in stopwords_list:
        clean_words.append(word)
print(f"Words without punication and stopwords is: {len(clean_words)}")
```

```python
# Generate and display a word cloud from the cleaned list of words
clean_words_string = " ".join(clean_words)
wordcloud = WordCloud(background_color = 'white').generate(clean_words_string)
plt.figure(figsize = (7,7))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
